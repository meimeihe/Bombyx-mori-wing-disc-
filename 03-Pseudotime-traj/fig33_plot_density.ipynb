```r
data<-read.csv(file="Pseudotime_cm.csv")
```

```r
data$timepoint<-factor(data$timepoint,levels = c("L5D1",
                 "L5D2",
                 "L5D3",
                 "L5D4",
                 "L5D5",
                 "L5D6",
                 "L5D7",
                 "WD1",
                 "WD2",
                 "P6"))
```

```r
options(repr.plot.width=10, repr.plot.height=6)  
wm = subset(data,celltype_v1 %in% c("Cell_morphogenesis"))
p_wm = ggplot(wm, aes(x = Pseudotime, y = celltype_v1,fill=timepoint)) +
   geom_density_ridges(aes(fill=timepoint), alpha =0.8,color = "white",from=0,to=40)+
  scale_fill_manual(values = pal_timepoint) +
  scale_y_discrete(expand = c(0.01, 0)) +
  scale_x_continuous(expand = c(0.01, 0))+labs( y = NULL) +
  # theme(plot.title = element_text(family, face, colour, size))
  theme_ridges()
  # theme()
p_wm
```
