```r
library(ggplot2)
library(dplyr)
library(tidyr) 
library(ggforce,lib.loc = "/hsfscqjf2/ST_CQ/Reference/software/envs/RTest/lib/R/library") 
```

```r
df_long<-as.data.frame(table(meta_data$timepoint,meta_data$celltype))
names(df_long)<-c("timepoint","celltype","count")
timepoint_levels <- c('0h', 
                   '10min', '20min', '30min', '40min', '50min', '60min', '4h', '6h','L5D1', 'L5D2', 'L5D3', 'L5D4', 'L5D5', 'L5D6', 'L5D7', 
                   'WD1', 'WD2', 'P6')
df_long$timepoint <- factor(df_long$timepoint, levels = timepoint_levels)
```

```r
head(df_long)
```

```r
order =c( "Ciliated_cell",
          "Axon_development",
          "Metabolic_process",
          "Apoptosis",
          "Immune",
          "Cuticle_3",
          "Cuticle_2",
          "Cuticle_1",
          "Epithelial_3",
          "Epithelial_2",
          "Epithelial_1",
          "Cell_morphogenesis"
)

df_long <- df_long %>%
  mutate(r_base = as.numeric(factor(celltype, levels = rev(order))))
```

```r
df_long <- df_long %>%
  group_by(celltype) %>%
  mutate(
    thickness = (count - min(count)) / (max(count) - min(count)) * 0.8 + 0.2 #
  ) %>%
  ungroup()
```

```r
df_long <- df_long %>%
  group_by(celltype) %>%
  mutate(
    timepoint_id = as.numeric(timepoint),
    sector_width = 2 * pi / n(), 
    start_angle = (timepoint_id - 1) * sector_width,
    end_angle = timepoint_id * sector_width
  ) %>%
  ungroup()
```

```r
head(df_long)
```

```r
p<-ggplot(df_long) +
  # 
  geom_arc_bar(
    aes(
      x0 = 0, y0 = 0,        
      r0 = r_base,           
      r = r_base + thickness, 
      start = start_angle, 
      end = end_angle,
      fill = timepoint       
    ),
    color = "white",         
    size = 0.2               
  ) +
  coord_fixed() +
  scale_fill_viridis_d(option = "viridis", name = "timepoint") +
  geom_text(
    data = ~ distinct(., celltype, r_base),
    aes(x = 0, 
        y = r_base + 0.1, # 
        label = celltype),
    hjust = 0.5,
    size = 2.5
  ) +
  theme_void() +
  theme(legend.position = "right") +
  labs(
    title = "Cell Count by Celltype and Timepoint",
    # subtitle = "Each ring represents a cell type. Ring thickness indicates normalized count at a timepoint.",
    fill = "timepoint"
  )
```

```r
options(repr.plot.width = 8, repr.plot.height=8)
print(p)
```
